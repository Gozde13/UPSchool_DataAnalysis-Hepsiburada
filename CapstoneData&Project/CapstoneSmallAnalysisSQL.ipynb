```python
# Let's get some of the datasets form Capstone Projects
# 6 datasets imported
customers = pd.read_csv('olist_customers_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_payments = pd.read_csv('olist_order_payments_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
```

```python
sql = '''
select product_category_name, product_id
from products
where product_category_name IS NOT NULL AND product_id IN

(select product_id
from order_items
where order_id IN 

(select order_id
from orders
where customer_id IN
(
select customer_id
from customers
where customer_unique_id ='8d50f5eadf50201ccdcedfb9e2ac8455'
)))

'''
pd.read_sql_query(sql, conn)
# Here we can see the product category name
# This customers buy sports leisure products a lot.
```

```python
#Let's start by looking six datasets 
sql = '''
select * 
from order_items
LIMIT 3
'''
pd.read_sql_query(sql, conn)
# order_id, order_item_id, product_id, seller_id, shipping_limit_date, price, freight_value
```

```python
sql = '''
select SUM(I.order_item_id), COUNT(O.order_id), O.customer_id, order_status,O.order_purchase_timestamp
from order_items I, orders O
where I.order_id=O.order_id 
group by customer_id
Order By order_item_id DESC

'''
pd.read_sql_query(sql, conn)
```

```python
# above and via this code we see that approximately 1 thousand customers purchased twice.
sql = '''
select C.customer_unique_id,SUM(I.order_item_id), COUNT(O.order_id), O.customer_id, order_status,O.order_purchase_timestamp
from order_items I, orders O, customers C
where I.order_id=O.order_id AND C.customer_id=O.customer_id
Group by customer_unique_id
Order By order_item_id DESC

'''
pd.read_sql_query(sql, conn)
```

# Now I want to look also payment values.


```python
sql = '''
select C.customer_unique_id,SUM(I.order_item_id) AS ItemsPurchased,COUNT(DISTINCT(I.product_id)) AS DistinctProduct, COUNT(O.order_id) AS OrderCounts
from order_items I, orders O, customers C
where I.order_id=O.order_id AND C.customer_id=O.customer_id
Group by customer_unique_id
Order By order_item_id DESC, OrderCounts DESC
limit 20
'''
pd.read_sql_query(sql, conn)
```

```python
# I wondered the count of payment installments in the data
type_versus_totalitem = pd.read_sql( '''
select payment_type,payment_installments,COUNT(I.order_id) AS Order_Count , SUM(order_item_id) AS total_items
from order_items I, order_payments P
where I.order_id=P.order_id AND payment_type='credit_card'
group by payment_installments
order by Order_Count DESC

''',conn)
type_versus_totalitem
# As seen below when credit card is used payment installemnts increasing.
# but it is seen also that Brazilians use mostly payment installment as 1.
```
